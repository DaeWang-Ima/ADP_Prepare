# 1. 포지셔닝 분석(Positioning Analysis)

- 포지셔닝 분석은 마케팅 통계분석 기법 중의 하나로, 기업이나, 상품, 브랜드 같은 개체들의 포지셔닝을 수행하는 다차원 척도법(Multi-Dimensional Scaling)과 상응(대응)분석(Correspondence Analysis)이 있다.


- 보유한 데이터가 주로 등간척도, 비율척도와 같이 구성되어 있다면 **다차원 척도법**을 사용한다.


- 보유한 데이터가 주로 명목척도, 서열척도와 같이 구성되어 있다면 **대응(상응)분석**을 사용한다.

# 2. 대응(상응) 분석(Correspondence Analysis)

- Correspondence Analysis를 통해 범주형 변수 범주들 사이의 관계를 살펴볼 수 있다.


- Correspondence Analysis는 다변량 범주형 자료를 대상으로 하는 탐색적 자료 분석이다.


- 이변량($X_1, X_2$)의 경우 단순대응분석이라고 하며 분할표의 정보를 저차원의 공간에 시각화하여 두 범주형 변수의 상관관계를 파악할 수 있다.


- 즉, 대응분석은 변수 간 상관관계 정보뿐만 아니라 변수가 갖는 범주들이 어떤 패턴으로 관련되어 있는지 알 수 있다는 장점이 있다.


- 대응분석을 수행하기 위해서는 빈도교차표를 생성해야 한다.


- **단순대응분석**은 2개의 변수를 **다중대응분석**은 3개 이상의 변수를 사용한다.

````python
# 라이브러리 불러오기
import pandas as pd

import warnings
warnings.filterwarnings(action = 'ignore')

import matplotlib.pyplot as plt
%matplotlib inline

# 대응분석을 수행할 수 있는 라이브러리 불러오기
import prince
````

## 2.1 Example - Hair Color and eye color

````python
# DataFrame 생성하기
hair_colors = ['Fair', 'Red', 'Medium', 'Dark', 'Black']
eye_colors = ['Blue', 'Light', 'Medium', 'Dark']
values = [[326, 38, 241, 110, 3],
          [688, 116, 584, 188, 4],
          [343, 84, 909, 412, 26],
          [98, 48, 403, 681, 85]]

data = pd.DataFrame(data = values, index = eye_colors, columns = hair_colors)
display(data)
````

- 제공되는 데이터가 없기 때문에 임의로 데이터를 생성했으며, 아래의 pandas 패키지의 내부 함수를 사용하여 쉽게 교차표를 생성할 수 있다.


```python
pd.crosstab(index = 인덱스로 사용할 변수, columns = 컬럼으로 사용할 변수, margins = False)
```

````python
# 대응분석 수행
ca = prince.CA(n_components = 2).fit(data)

# 시각화 수행
ax = ca.plot_coordinates(X = data, figsize = (12, 6));
````
