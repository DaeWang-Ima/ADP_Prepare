## 1.1 서론

- 로지스틱 회귀는 선형 회귀 방식을 분류에 적용한 알고리즘이며, 종속변수 Y가 범주형인 경우로 확장한 것이다.


- 로지스틱 회귀모델은 다양한 분야에서, 반응변수를 설명하거나 예측하기 위해 구조화된 모델이 필요할 때마다 사용된다.


- 로지스틱 회귀는 예측변수의 값을 바탕으로 클래스가 알려져 있지 않은 새로운 관측치를 클래스들 중 하나로 보낼 때 사용될 수 있다.


- 또한, **클래스가 알려진 데이터**에서는, 예측변수의 관점에서 서로 다른 클래스의 관측치들을 구별하는 요인을 찾을 때 사용될 수도 있다. 


> **로지스틱 회귀분석 응용분야**
>> 1. 고객을 재구매 고객과 처음 구매한 고객으로 분류 
>> 2. 남자 최고 경영진과 여자 최고 경영진을 구별하는 요인 찾기
>> 3. 신용점수와 같은 정보로부터 대출의 승인 또는 비승인 예측


- 로지스틱 회귀분석의 예측변수 $X_1, X_2, ..., X_k$는 범주형 변수이거나 연속형 변수, 또는 이들 두 가지를 혼합한 형태가 올 수 있다.


- 로지스틱 회귀는 다음의 **두 단계**로 이루어졌다.

    1. 첫 번째 단계는 각 클래스에 속하는 **성향(Propensities)** 또는 **확률(Probabilities)** 을 추정하는 것이다. 이진형 변수인 경우 클래스 1에 속할 확률인 $p=P(Y=1)$을 추정한다.

    2. 두 번째 단계는 각 관측치의 클래스를 지정하기 위해 확률값에 대한 **컷오프(Cutoff)** 값을 적용하는 것이다. 예를 들어, 이진형 변수의 경우에 컷오프가 0.5라면 P(Y=1)의 추정확률이 0.5 이상일 때 클래스 1로, 추정확률이 0.5보다 작으면 클래스 0으로 분류된다. 이때 컷오프를 0.5로 고정할 필요가 없으며, 해당 사건의 발생가능성이 낮지만 중요한 사건인 경우에는, 더 많은 관측치들이 클래스 1에 속하도록 하기 위해 낮은 컷오프가 사용될 수 있다.

## 1.2 로지스틱 회귀모델

- 로지스틱 회귀의 원리는 종속변수 Y를 대신해서 **로짓(Logit)** 이라고 부르는 Y의 함수를 사용하는 것이다.


- 로짓을 예측변수들의 선형함수로 모형화하는 것이며, 로짓이 예측되면 그로부터 확률을 다시 구할 수 있다.


- 로짓을 이해하기 위해서는 몇 가지 중간단계를 거쳐야 하며, 그 과정들을 자세히 살펴보도록 하자.


> 우선 이진형 종속변수로 가정하고 첫 번째로 클래스 1에 속할 확률 $p$를 구한다. 클래스 1에 속할 확률이 $p$라면, 클래스 0에 속할 확률은 $1-p$가 된다.<br><br>
0과 1 두 개의 값만 가질수 있는 Y와 달리 $p$는 구간 [0, 1]에 존재하는 모든 값을 가질 수 있다. 확률 $p$를 $q$개의 예측변수들에 대한 선형함수로 나타내면 아래의 식과 같다.<br><br>
$$p=\beta_0 + \beta_{1}x_1 + \beta_{2}x_2 + ... + \beta_{q}x_q$$<br>
위의 식에서 $p$는 구간 [0, 1]에 들어가지만, 우변항이 구간 [0, 1]에 들어간다는 것을 보장할 수 없게 된다. 이러한 문제를 해결하기 위해 다음과 같이 예측변수들에 대한 비선형 함수를 사용하게 된다.<br><br>
$$p=\frac{1}{1+e^{-(\beta_0 + \beta_{1}x_1 + \beta_{2}x_2 + ... + \beta_{q}x_q)}}$$<br>
임의의 예측변수 $x_1, ..., x_q$에 대해 우변은 항상 구간 [0, 1] 사이의 값이 된다.<br><br>
다음으로 <span style="color:yellow">**오즈(Odds)**</span> 에 대해서 알아보도록 하자. 오즈란 어떤 사건이 일어날 확률과 일어나지 않을 확률의 비를 의미한다. 즉, 클래스 1에 속할 오즈는 **클래스 0에 속하는 확률에 대한 클래스 1에 속하는 확률의 비**로 정의된다.<br><br>
$$Odds(Y=1)=\frac{p}{1-p}$$<br>
예를 들어, 병에 걸릴 확률이 0.5일 때 병에 걸릴 오즈는 $\frac{0.5}{1-0.5}=1$이다. 역으로 오즈로부터 확률을 계산할 수 있는데, 바로 위의 식을 변형하면 된다.<br><br>
$$p=\frac{Odds}{1+Odds}$$<br>
이때 두 번째 식을 네 번쨰 식에 대입하면, 예측변수와 오즈의 관계는 다음과 같이 정의된다.<br><br>
$$Odds(Y=1)=e^{\beta_0 + \beta_{1}x_1 + \beta_{2}x_2 + ... + \beta_{q}x_q}$$<br>
이제 위의 식 양변에 자연로그를 취하면 로지스틱 모델의 표준형을 얻게 된다.<br><br>
$$log(Odds)=\beta_0 + \beta_{1}x_1 + \beta_{2}x_2 + ... + \beta_{q}x_q$$<br>
이때 우리는 log(Odds)를 <span style="color:yellow">**로짓(Logit)**</span> 이라고 부르며, $-\infty$부터 $\infty$ 사이의 값을 가질 수 있다. 따라서 반응변수와 예측변수의 관계를 나타내는 최종형태는 로짓을 종속변수로 하여 q개의 예측변수에 대한 선형 함수로 모형화하는 것이다.
